```python
import pandas as pd
import numpy as np


import geopandas 
import folium
from folium.plugins import MarkerCluster


import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path_data = "../data/datos/CM03/"
```

### Mapa

```python
path_islas = "https://datos.canarias.es/catalogos/estadisticas/dataset/24d0ed91-1ce1-47e3-8b4c-44cf843fec28/resource/b3f12117-e4fd-4b8c-a31f-a93a1a8a020d/download/islas_20170101.json"
canaria = geopandas.read_file(path_islas)
```

### Gastos de hoteles

```python
plan = pd.read_csv(path_data+"maestro_cm_03_comportamiento_islas.csv")
plan.loc[:, "fecha"] = pd.to_datetime(plan["fecha"], format="%Y%m%d")
plan.head()
```

```python
for i in ['indicador', 'dimension', 'valor_dimension']:
    print("Los valores unicos de "+i+": ")
    print(plan[i].unique().tolist())
    print("---------------------------------------------")
```

```python
plan_totales = plan[plan["valor_dimension"]=="Total"]
```

```python
for i in plan_totales["indicador"].unique():
    fig = plt.figure(figsize=(16,4))
    sns.lineplot(data=plan_totales[plan_totales["indicador"]==i], x="fecha", y="valor", hue="isla")
    plt.xticks(rotation=90)
    plt.title(i)
    plt.show()
```

### Gran Canaria

```python
plan_canaria = plan[(plan.isla=="Gran Canaria") & (plan["valor_dimension"]!="Total")]

indicador = "Gasto por turista"
# indicador = "Gasto por turista y día"

for i in plan_canaria["dimension"].unique():
    print(i)
    fig = plt.figure(figsize=(16,4))
    sns.lineplot(data=plan_canaria[(plan_canaria["indicador"]==indicador) & (plan_canaria["dimension"]==i)], 
                 x="fecha", y="valor", hue="valor_dimension")
    #plt.legend([])
    plt.xticks(rotation=90)
    plt.title(i)
    plt.show()
```

## Gasto por edad y sexo 

```python
edad_sexo = pd.read_csv(path_data+"maestro_cm_03_demograficas_islas.csv")
edad_sexo["fecha"] = pd.to_datetime(edad_sexo["fecha"], format="%Y%m%d")
edad_sexo.head()
```

```python
plan_canaria = edad_sexo[(edad_sexo.isla=="Gran Canaria") & (~edad_sexo["valor_dimension"].isin(["Ambos sexos", 
                                                                                                "Total grupos de edades"]))]

indicador = "Gasto por turista"
# indicador = "Gasto por turista y día"

for i in plan_canaria["dimension"].unique():
    print(i)
    fig = plt.figure(figsize=(16,4))
    sns.lineplot(data=plan_canaria[(plan_canaria["indicador"]==indicador) & (plan_canaria["dimension"]==i)], 
                 x="fecha", y="valor", hue="valor_dimension")
    #plt.legend([])
    plt.xticks(rotation=90)
    plt.title(i)
    plt.show()
```

## Situacion laboral, estudios, nivel de ingresos

```python
perfil = pd.read_csv(path_data+"maestro_cm_03_socio_economicas_islas.csv")
perfil["fecha"] = pd.to_datetime(perfil["fecha"], format="%Y%m%d")
perfil.head()

plan_canaria = perfil[(perfil.isla=="Gran Canaria") & (~perfil["valor_dimension"].isin(["Total", 
                                                                                                "Total grupos de edades"]))]

#indicador = "Gasto por turista"
indicador = "Gasto por turista y día"

for i in plan_canaria["dimension"].unique():
    print(i)
    fig = plt.figure(figsize=(16,4))
    sns.lineplot(data=plan_canaria[(plan_canaria["indicador"]==indicador) & (plan_canaria["dimension"]==i)], 
                 x="fecha", y="valor", hue="valor_dimension")
    #plt.legend([])
    plt.xticks(rotation=90)
    plt.title(i)
    plt.show()

```
