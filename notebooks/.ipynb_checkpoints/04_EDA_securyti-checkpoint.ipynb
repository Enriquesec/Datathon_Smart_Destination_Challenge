# EDA

```python
import os 
import glob


import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
securyty = pd.read_csv(path+"maestro_cm35.csv")
securyty = securyty[securyty.provincia=="CANARIAS"].copy()
securyty.head()
```

```python
for i in ['indicador', "tipo_infraccion", "nacionalidad"]:
    print("Los valores unicos de "+i+": ")
    print(securyty[i].unique().tolist())
    print("---------------------------------------------")
```

```python
for indica in securyty.indicador.unique().tolist():
    print(indica)
    for j in ["tipo_infraccion", "nacionalidad"]:
        print("Los valores unicos de "+j+": ")
        print(securyty.loc[(securyty.indicador==indica),j].unique().tolist())
    print("---------------------------------------------")
```
