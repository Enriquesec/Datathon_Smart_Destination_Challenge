# EDA

```python
import os 
import glob


import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
maestro = pd.read_csv(path+"maestro_cm21.csv")
maestro = maestro[maestro.lugar=="Isla De Gran Canaria"]
maestro.head()
```

```python
for i in ['origen', "indicador", "dimension"]:
    print("Los valores unicos de "+i+": ")
    print(maestro[i].unique().tolist())
    print("---------------------------------------------")
```

```python
maestro.info()
```

```python
maestro["mes"]= maestro.mes.map("{:02}".format)
maestro["fecha"] = pd.to_datetime(maestro["anyo"].astype(str)+maestro["mes"].astype(str)+"01", 
                                 format="%Y%m%d")

maestro.origen.fillna("Others", inplace=True)
```

```python
maestro
```

```python
for indicador in maestro.indicador.unique():
    aux = maestro[maestro.indicador==indicador]
    fig = plt.figure(figsize=(12,6))
    sns.lineplot(data=aux, x="fecha", y="valor", hue="origen")
    plt.title(indicador)
    plt.show()
```
